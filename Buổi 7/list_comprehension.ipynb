# List Comprehension trong Python

**Cú pháp cơ bản:**
```python
[expression for item in iterable if condition]
```

List comprehension giúp tạo list ngắn gọn, dễ đọc hơn so với vòng lặp `for` thông thường.

## 1. Cơ bản — Bình phương các số

````python
# Cách thông thường
squares_loop = []
for x in range(1, 11):
    squares_loop.append(x ** 2)

# List comprehension
squares = [x ** 2 for x in range(1, 11)]

print(squares)
````

## 2. Lọc phần tử — Số chẵn trong danh sách

````python
numbers = range(1, 21)

evens = [n for n in numbers if n % 2 == 0]

print(evens)
````

## 3. Biến đổi chuỗi — Chuyển chữ hoa/thường

````python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']

upper_fruits = [fruit.upper() for fruit in fruits]
long_fruits   = [fruit.upper() for fruit in fruits if len(fruit) > 5]

print(upper_fruits)
print(long_fruits)
````

## 4. Kết hợp điều kiện `if-else` (ternary)

````python
numbers = range(1, 11)

# Thay số lẻ bằng 0
result = [n if n % 2 == 0 else 0 for n in numbers]

print(result)
````

## 5. Làm phẳng ma trận 2D (flatten) lồng 2 vòng lặp

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flat = [num for row in matrix for num in row]

print(flat)
````

## 6. Nested list comprehension — Tạo ma trận nhân

````python
# Bảng nhân 3x3
table = [[i * j for j in range(1, 4)] for i in range(1, 4)]
print(table)
for row in table:
    print(row)
````

## 7. Lọc và biến đổi dict — Tìm học sinh đạt

````python
students = {
    'An': 8.5,
    'Bình': 5.0,
    'Chi': 9.2,
    'Dũng': 4.8,
    'Em': 7.3
}

# Tên học sinh đạt (điểm >= 6.5)
passed = [name for name, score in students.items() if score >= 6.5]

# Dict mới chỉ gồm học sinh đạt
passed_dict = {name: score for name, score in students.items() if score >= 6.5}

print('Đạt:', passed)
print('Chi tiết:', passed_dict)
````

## 8. Loại bỏ phần tử trùng lặp (giữ thứ tự)

````python
data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

seen = set()
unique = [x for x in data if not (x in seen or seen.add(x))]

print(unique)
````

## 9. Tách từ và lọc stop words

````python
sentence  = "the quick brown fox jumps over the lazy dog"
stopwords = {'the', 'over', 'a', 'an'}

words = [word for word in sentence.split() if word not in stopwords]

print(words)
````

## 10. Set comprehension & Dict comprehension

````python
numbers = [1, 2, 2, 3, 4, 4, 5]

# Set comprehension — tự động loại trùng
unique_squares = {x ** 2 for x in numbers}

# Dict comprehension — số → bình phương
square_map = {x: x ** 2 for x in range(1, 6)}

print('Set  :', unique_squares)
print('Dict :', square_map)
````

## 11. Generator expression — Tiết kiệm bộ nhớ

````python
# Thay [] bằng () → generator (lazy evaluation)
gen = (x ** 2 for x in range(1, 1_000_001))

# Không tạo toàn bộ list trong bộ nhớ
total = sum(gen)

print(f'Tổng bình phương 1..1000000 = {total:,}')
````

## 12. Bài tập tổng hợp — FizzBuzz

````python
fizzbuzz = [
    'FizzBuzz' 
    if n % 15 == 0
    else 'Fizz'  if n % 3  == 0
    else 'Buzz'  if n % 5  == 0
    else str(n)
    for n in range(1, 31)
]

print(fizzbuzz)
````
