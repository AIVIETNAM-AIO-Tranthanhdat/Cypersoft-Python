```python
# Class & Object
class Dog:
    species = "Canis familiaris"      # class attribute (chung cho mọi instance)

    def __init__(self, name, age):    # constructor
        self.name = name              # instance attribute
        self.age = age

    def bark(self):                   # method chỉ có parameter là chính nó
        return f"{self.name} says Woof!"
    
    def theloai(self,theloai):        # method có nhiều parameter
        if theloai == "tây":
            print ('Hãy cho ăn nhiều steak')
    

dog1 = Dog("Rex", 3)# tạo object (instance)
print (dog1.bark()) # Truy cập method 01 của đối tượng
print (dog1.theloai('tây')) # Truy cập method 02 của đối tượng
print (dog1.name) # Truy cập thuộc tính 01 của đối tượng
print (dog1.age) # Truy cập thuộc tính 02 của đối tượng 
```
