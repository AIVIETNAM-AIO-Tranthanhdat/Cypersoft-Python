##**Bài tập 1:**

Yêu cầu:
1. Tạo class SinhVien với __init__(ten, tuoi, lop)
2. Nhập thông tin 2 sinh viên từ bàn phím
3. Tạo 2 object từ dữ liệu nhập
4. In thông tin cả 2 SV

##**Bài tập 3:**
Yêu cầu:
1. Tạo class TaiKhoan với __init__(ten, so_du)
2. Thêm method nap_tien(so_tien):
   - Cộng so_tien vào self.so_du
   - In "Nạp [so_tien] thành công. Số dư: [so_du]"
3. Thêm method rut_tien(so_tien):
   - Nếu đủ tiền: trừ tiền, in thông báo
   - Nếu không đủ: in "Số dư không đủ!"
4. Tạo object, thử nạp và rút

```python
class taikhoan:
    def __init__ (self,name,numbers):
        self.name = name
        self.numbers = numbers
    def nap_tien(self,sotien):
        self.numbers = self.numbers + sotien
        print (f'Bạn đã nạp tiền thành công. Giao dịch: {sotien}. Số dư: {self.numbers}')
    def rut_tien(self,sotien):
        if self.numbers >= sotien:
            self.numbers = self.numbers - sotien
            print (f"Bạn đã rút thành công: {sotien}")
        else:
            print (f'Số dư không đủ')
taikhoan01 = taikhoan ('Đạt',200000)
taikhoan01.nap_tien(300000)
print (taikhoan01.numbers)
taikhoan01.rut_tien(500000)
print (taikhoan01.numbers)
taikhoan01.rut_tien(100)
print (taikhoan01.numbers)
```

##**Bài tập 4:**
Yêu cầu:
1. Tạo class HinhTron với __init__(ban_kinh)
2. Thêm method tinh_chu_vi() - return 2 * 3.14 * ban_kinh
3. Thêm method tinh_dien_tich() - return 3.14 * ban_kinh ** 2
4. Tạo object, gọi 2 method, in kết quả

```python
class HinhTron:
    def __init__(self, ban_kinh):
        self.ban_kinh = ban_kinh

    def tinh_chu_vi(self):
        return 2 * 3.14 * self.ban_kinh

    def tinh_dien_tich(self):
        return 3.14 * self.ban_kinh ** 2

ht = HinhTron(5)
print(f"Chu vi: {ht.tinh_chu_vi()}")
print(f"Diện tích: {ht.tinh_dien_tich()}")
```
